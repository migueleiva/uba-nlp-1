NГіtese que estamos estructurando el problema de aprendizaje como *many-to-many*:

Entrada: secuencia de tokens [$x_0$, $x_1$, ..., $x_N$]

Target: secuencia de tokens [$x_1$, $x_2$, ..., $x_{N+1}$]

De manera que la red tiene que aprender que su salida deben ser los tokens desplazados en una posiciГіn y un nuevo token predicho (el N+1).

La ventaja de estructurar el aprendizaje de esta manera es que para cada token de target se propaga una seГ±al de gradiente por el grafo de cГіmputo recurrente, que es mejor que estructurar el problema como *many-to-one* en donde sГіlo una seГ±al de gradiente se propaga.


### Definir el modelo

Dado que por el momento no hay implementaciones adecuadas de la perplejidad que puedan operar en tiempo de entrenamiento, armaremos un Callback *ad-hoc* que la calcule en cada epoch.

**Nota**: un Callback es una rutina gatillada por algГєn evento, son muy Гєtiles para relevar datos en diferentes momentos del desarrollo del modelo. En este caso queremos hacer un cГЎlculo cada vez que termina una epoch de entrenamiento.

```python
class PplCallback(keras.callbacks.Callback):

    '''
    Este callback es una soluciГіn ad-hoc para calcular al final de cada epoch de
    entrenamiento la mГ©trica de Perplejidad sobre un conjunto de datos de validaciГіn.
    La perplejidad es una mГ©trica cuantitativa para evaluar la calidad de la generaciГіn de secuencias.
    AdemГЎs implementa la finalizaciГіn del entrenamiento (Early Stopping)
    si la perplejidad no mejora despuГ©s de `patience` epochs.
    '''

    def __init__(self, val_data, history_ppl,patience=5):
      # El callback lo inicializamos con secuencias de validaciГіn sobre las cuales
      # mediremos la perplejidad
      self.val_data = val_data

      self.target = []
      self.padded = []

      count = 0
      self.info = []
      self.min_score = np.inf
      self.patience_counter = 0
      self.patience = patience

      # nos movemos en todas las secuencias de los datos de validaciГіn
      for seq in self.val_data:

        len_seq = len(seq)
        # armamos todas las subsecuencias
        subseq = [seq[:i] for i in range(1,len_seq)]
        self.target.extend([seq[i] for i in range(1,len_seq)])

        if len(subseq)!=0:

          self.padded.append(pad_sequences(subseq, maxlen=max_context_size, padding='pre'))

          self.info.append((count,count+len_seq))
          count += len_seq

      self.padded = np.vstack(self.padded)


    def on_epoch_end(self, epoch, logs=None):

        # en `scores` iremos guardando la perplejidad de cada secuencia
        scores = []

        predictions = self.model.predict(self.padded,verbose=0)

        # para cada secuencia de validaciГіn
        for start,end in self.info:

          # en `probs` iremos guardando las probabilidades de los tГ©rminos target
          probs = [predictions[idx_seq,-1,idx_vocab] for idx_seq, idx_vocab in zip(range(start,end),self.target[start:end])]

          # calculamos la perplejidad por medio de logaritmos
          scores.append(np.exp(-np.sum(np.log(probs))/(end-start)))

        # promediamos todos los scores e imprimimos el valor promedio
        current_score = np.mean(scores)
        history_ppl.append(current_score)
        print(f'\n mean perplexity: {current_score} \n')

        # chequeamos si tenemos que detener el entrenamiento
        if current_score < self.min_score:
          self.min_score = current_score
          self.model.save("my_model.keras")
          print("Saved new model!")
          self.patience_counter = 0
        else:
          self.patience_counter += 1
          if self.patience_counter == self.patience:
            print("Stopping training...")
            self.model.stop_training = True
```

### GeneraciГіn de secuencias

```python
def generate_seq(model, seed_text, max_length, n_words):
    """
        Exec model sequence prediction

        Args:
            model (keras): modelo entrenado
            seed_text (string): texto de entrada (input_seq)
            max_length (int): mГЎxima longitud de la sequencia de entrada
            n_words (int): nГєmeros de caracteres a agregar a la sequencia de entrada
        returns:
            output_text (string): sentencia con las "n_words" agregadas
    """
    output_text = seed_text
	# generate a fixed number of words
    for _ in range(n_words):
		# Encodeamos
        encoded = [char2idx[ch] for ch in output_text.lower() ]
		# Si tienen distinto largo
        encoded = pad_sequences([encoded], maxlen=max_length, padding='pre')

		# PredicciГіn softmax
        y_hat = np.argmax(model.predict(encoded,verbose=0)[0,-1,:])
		# Vamos concatenando las predicciones
        out_word = ''

        out_word = idx2char[y_hat]

		# Agrego las palabras a la frase predicha
        output_text += out_word
    return output_text
```

###  Beam search y muestreo aleatorio

```python
# funcionalidades para hacer encoding y decoding

def encode(text,max_length=max_context_size):

    encoded = [char2idx[ch] for ch in text]
    encoded = pad_sequences([encoded], maxlen=max_length, padding='pre')

    return encoded

def decode(seq):
    return ''.join([idx2char[ch] for ch in seq])
```

```python
from scipy.special import softmax

# funciГіn que selecciona candidatos para el beam search
def select_candidates(pred,num_beams,vocab_size,history_probs,history_tokens,temp,mode):

  # colectar todas las probabilidades para la siguiente bГєsqueda
  pred_large = []

  for idx,pp in enumerate(pred):
    pred_large.extend(np.log(pp+1E-10)+history_probs[idx])

  pred_large = np.array(pred_large)

  # criterio de selecciГіn
  if mode == 'det':
    idx_select = np.argsort(pred_large)[::-1][:num_beams] # beam search determinista
  elif mode == 'sto':
    idx_select = np.random.choice(np.arange(pred_large.shape[0]), num_beams, p=softmax(pred_large/temp)) # beam search con muestreo aleatorio
  else:
    raise ValueError(f'Wrong selection mode. {mode} was given. det and sto are supported.')

  # traducir a Г­ndices de token en el vocabulario
  new_history_tokens = np.concatenate((np.array(history_tokens)[idx_select//vocab_size],
                        np.array([idx_select%vocab_size]).T),
                      axis=1)

  # devolver el producto de las probabilidades (log) y la secuencia de tokens seleccionados
  return pred_large[idx_select.astype(int)], new_history_tokens.astype(int)


def beam_search(model,num_beams,num_words,input,temp=1,mode='det'):

    # first iteration

    # encode
    encoded = encode(input)

    # first prediction
    y_hat = model.predict(encoded,verbose=0)[0,-1,:]

    # get vocabulary size
    vocab_size = y_hat.shape[0]

    # initialize history
    history_probs = [0]*num_beams
    history_tokens = [encoded[0]]*num_beams

    # select num_beams candidates
    history_probs, history_tokens = select_candidates([y_hat],
                                        num_beams,
                                        vocab_size,
                                        history_probs,
                                        history_tokens,
                                        temp,
                                        mode)

    # beam search loop
    for i in range(num_words-1):

      preds = []

      for hist in history_tokens:

        # actualizar secuencia de tokens
        input_update = np.array([hist[i+1:]]).copy()

        # predicciГіn
        y_hat = model.predict(input_update,verbose=0)[0,-1,:]

        preds.append(y_hat)

      history_probs, history_tokens = select_candidates(preds,
                                                        num_beams,
                                                        vocab_size,
                                                        history_probs,
                                                        history_tokens,
                                                        temp,
                                                        mode)

    return history_tokens[:,-(len(input)+num_words):]
```

### Paso 2: TokenizaciГіn Moderna con Keras (TextVectorization)

```python
import tensorflow as tf
import re

def custom_standardization(input_data):
    lowercase = tf.strings.lower(input_data)

    return lowercase

max_context_size = 100

vectorizer = tf.keras.layers.TextVectorization(
    split="character",
    standardize=custom_standardization 
)

vectorizer.adapt([raw_text]) 
vocab = vectorizer.get_vocabulary()
vocab_size = len(vocab)

print(f"TamaГ±o del vocabulario Гєnico encontrados: {vocab_size}")
print("Primeros 10 tokens en el vocabulario:", vocab[:10])

encoded_text = vectorizer([raw_text])[0].numpy()
```

### Paso 3: EstructuraciГіn del Dataset (Many-to-Many)

```python
# Reservamos el 10% final del texto para validaciГіn
p_val = 0.1
num_val_chars = int(len(encoded_text) * p_val)

train_text = encoded_text[:-num_val_chars]
val_text = encoded_text[-num_val_chars:]

# Creamos secuencias de tamaГ±o max_context_size deslizando la ventana de 1 en 1
# Para entrenamiento (X_train) y el target desfasado un caracter (y_train)
X_train_list = [train_text[i : i + max_context_size] for i in range(len(train_text) - max_context_size)]
y_train_list = [train_text[i + 1 : i + max_context_size + 1] for i in range(len(train_text) - max_context_size)]

X_train = np.array(X_train_list)
y_train = np.array(y_train_list)

# Hacemos exactamente lo mismo para el set de validaciГіn
X_val_list = [val_text[i : i + max_context_size] for i in range(len(val_text) - max_context_size)]
y_val_list = [val_text[i + 1 : i + max_context_size + 1] for i in range(len(val_text) - max_context_size)]

X_val = np.array(X_val_list)
y_val = np.array(y_val_list)

print(f"Dimensiones de X_train (Secuencias, TamaГ±o Contexto): {X_train.shape}")
print(f"Dimensiones de y_train: {y_train.shape}")
```

### Paso 4: DefiniciГіn de las 3 Arquitecturas Recurrentes
Para mantener el cГіdigo simple de entender, creo una funciГіn genГ©rica que construya el modelo (SimpleRNN, LSTM o GRU).

```python
def build_language_model(rnn_layer_type, vocab_size, embedding_dim=64, rnn_units=256):
    model = keras.Sequential()
    model.add(layers.Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_shape=(None,)))
    
    if rnn_layer_type == 'SimpleRNN':
        model.add(layers.SimpleRNN(rnn_units, return_sequences=True, dropout=0.2))
    elif rnn_layer_type == 'LSTM':
        model.add(layers.LSTM(rnn_units, return_sequences=True, dropout=0.2))
    elif rnn_layer_type == 'GRU':
        model.add(layers.GRU(rnn_units, return_sequences=True, dropout=0.2))
        
    model.add(layers.Dense(vocab_size, activation='softmax'))
    
    model.compile(optimizer='rmsprop', loss='sparse_categorical_crossentropy')
    return model
```

```python
model_simple_rnn = build_language_model('SimpleRNN', vocab_size)
model_lstm = build_language_model('LSTM', vocab_size)
model_gru = build_language_model('GRU', vocab_size)
```

### Paso 5: Callback de Perplejidad Simplificado y Entrenamiento
Utilizaremos el mГ©todo directo (calcular el exponente de la pГ©rdida en validaciГіn), guardando el historial de cada modelo para compararlos al final.

```python
class PerplexityCallback(keras.callbacks.Callback):
    def __init__(self, model_name):
        super().__init__()
        self.model_name = model_name
        self.history_ppl = []
        self.min_ppl = np.inf
        self.patience_counter = 0
        self.patience = 4 

    def on_epoch_end(self, epoch, logs=None):
        val_loss = logs.get('val_loss')
        perplexity = np.exp(val_loss)
        self.history_ppl.append(perplexity)
        print(f"\n[{self.model_name}] - Perplejidad en ValidaciГіn: {perplexity:.2f}")
        
        if perplexity < self.min_ppl:
            self.min_ppl = perplexity
            self.model.save(f"{self.model_name}_best.keras")
            self.patience_counter = 0
        else:
            self.patience_counter += 1
            if self.patience_counter >= self.patience:
                print("ВЎFrenado temprano activado por falta de mejora en Perplejidad!")
                self.model.stop_training = True

callbacks = {
    'SimpleRNN': PerplexityCallback('SimpleRNN'),
    'LSTM': PerplexityCallback('LSTM'),
    'GRU': PerplexityCallback('GRU')
}
```

```python
print("=== Entrenando SimpleRNN ===")
model_simple_rnn.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=10, batch_size=256, callbacks=[callbacks['SimpleRNN']])

print("\n=== Entrenando LSTM ===")
model_lstm.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=10, batch_size=256, callbacks=[callbacks['LSTM']])

print("\n=== Entrenando GRU ===")
model_gru.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=10, batch_size=256, callbacks=[callbacks['GRU']])
```

### A) Greedy Search (Estrategia Codiciosa)
Toma siempre el carГЎcter con la probabilidad mГЎs alta (argmax).

```python
def generate_greedy(model, seed_text, n_chars=100):
    output_text = seed_text
    for _ in range(n_chars):
        encoded = text_to_indices(output_text)
        encoded = pad_sequences([encoded], maxlen=max_context_size, padding='pre')
        
        preds = model.predict(encoded, verbose=0)[0, -1, :]
        next_index = np.argmax(preds)
        
        output_text += vocab[next_index]
    return output_text

print("Greedy RNN:", generate_greedy(best_rnn, "Mas tambiГ©n quiero que sepa vuestra "))
print("Greedy LSTM:", generate_greedy(best_lstm, "Mas tambiГ©n quiero que sepa vuestra "))
print("Greedy GRU:", generate_greedy(best_gru, "Mas tambiГ©n quiero que sepa vuestra "))
```

### B) Beam Search DeterminГ­sticoMantiene las $K$ mejores hipГіtesis vivas en cada paso de tiempo de forma matemГЎtica rГ­gida.

```python
def generate_beam_search_det(model, seed_text, n_chars=50, beam_width=3):
    candidates = [(seed_text, 0.0)]
    
    for _ in range(n_chars):
        all_successors = []
        for text_seq, score in candidates:
            encoded = text_to_indices(text_seq)
            encoded = pad_sequences([encoded], maxlen=max_context_size, padding='pre')
            
            preds = model.predict(encoded, verbose=0)[0, -1, :]
            
            top_k_indices = np.argsort(preds)[-beam_width:]
            for idx in top_k_indices:
                next_char = vocab[idx]
                new_score = score - np.log(preds[idx] + 1e-10)
                all_successors.append((text_seq + next_char, new_score))
                
        all_successors = sorted(all_successors, key=lambda x: x[1])
        candidates = all_successors[:beam_width]
        
    return candidates[0][0]

print("Beam Search DeterminГ­stico (GRU):", generate_beam_search_det(best_gru, "Mas tambiГ©n quiero que sepa vuestra ", n_chars=60, beam_width=4))
```

#### C) Beam Search EstocГЎstico (Muestreo Aleatorio con Temperatura)
Modifique las probabilidades usando la temperatura y luego "sortea" (hace un sampleo aleatorio) cuГЎl serГЎ el siguiente carГЎcter.

```python
def generate_stochastic_temp(model, seed_text, n_chars=100, temperature=1.0):
    output_text = seed_text
    for _ in range(n_chars):
        encoded = text_to_indices(output_text)
        encoded = pad_sequences([encoded], maxlen=max_context_size, padding='pre')
        
        preds = model.predict(encoded, verbose=0)[0, -1, :]
        log_preds = np.log(preds + 1e-10) / temperature
        exp_preds = np.exp(log_preds)
        adjusted_probs = exp_preds / np.sum(exp_preds)
        
        next_index = np.random.choice(len(adjusted_probs), p=adjusted_probs)
        output_text += vocab[next_index]
        
    return output_text
```
